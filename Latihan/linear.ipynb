```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0, 20)

a = 1
b = 2

# linear growth
# f(x) = 2x + 1
# linear = b*t + a

# exponential growth
# f(x) = 2^x
expo = b * np.exp(t)

# plotting
# plt.plot(t, linear, label="Linear")

plt.xlabel("Time")
plt.ylabel("Bacteria Count")
plt.plot(t, expo, label="Exponential")
plt.legend()
plt.show()
```
