```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data
x = np.linspace(0,5,100)
y = [2**i for i in x]
y2 = [3**i for i in x]
# Membuat Plot garis
plt.plot(x,y, label="2^x")
plt.plot(x,y2, label="3^x")

# Menambahkan judul dan label Sumbu
plt.title("Plot Fungsi Eksponen f(x) = 2^x")
plt.xlabel("Sumbu X")
plt.ylabel("Sumbu Y")
plt.legend()
plt.grid(True)
# Menampilkan Plot
plt.show()
```
